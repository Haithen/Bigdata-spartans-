```python
from kafka import KafkaProducer
import csv
```

```python
# Kafka producer configuration
kafka_bootstrap_servers = '34.215.164.92:9092'
kafka_topic = 'lakers-zillow'
```

```python
producer = KafkaProducer(bootstrap_servers=kafka_bootstrap_servers)
```

```python
csv_file_path = 'lakers-zillow.csv'
```

```python
import pandas as pd
pd.read_csv(csv_file_path)
```

```python
with open(csv_file_path, 'r') as file:
    for line in file:
        # Publish each line as a Kafka message
        producer.send(kafka_topic, value=line.encode('utf-8'))
```

```python
producer.flush()
producer.close()
```

```python
producer.close()
```
